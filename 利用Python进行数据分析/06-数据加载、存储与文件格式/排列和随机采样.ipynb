#### 利用numpy.random.permutation函数可以随机重排序，
#### 通过需要排列的轴的长度调用permutation。可产生一个新顺序的整数数组

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.arange(5*4).reshape((5,4)))
sampler = np.random.permutation(df.shape[0])
sampler
```

```python
df.iloc[sampler]
```

```python
df.sample(n =3)
```

### 指标和哑变量 get_dummies
> 将分类变量转换成为“哑变量”或者“指标矩阵”

```python
df = pd.DataFrame({'key':['b','b','a','c','a','b'],
                  'data1':range(6)})
pd.get_dummies(df['key'])
```

```python
dumplies = pd.get_dummies(df['key'],prefix = '分类')
dumplies
```

```python
df_with_dumplies = df.join(dumplies)
df_with_dumplies
```

```python
np.random.seed(12345)
values = np.random.rand(10)
values
```

```python
bins = [0,0.2,0.4,0.6,0.8,1]
pd.get_dummies(pd.cut(values,bins))
```
