```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
# 自己创建数据集
x,y = make_blobs(n_samples = 1000,n_features = 2,centers =4,random_state = 1)

x # 表示的数据集，简单的可以理解为横纵坐标
y # 表示的是数据集所处的是哪一类，因为我们的centers设定为4，所以有0123的分类，
  #对于n_features=2则说明这是一个2维的数据集
```

```python
fig,ax1 = plt.subplots(1)
ax1.scatter(x[:,0],x[:,1],
           marker = 'o', # 点的形状
           s = 5 # 点的大小
           )
plt.show()
```

```python
# 利用y的值分别进行分类，画出分类后的图形
colors = ['black','red','yellow','blue']
fig,ax1 = plt.subplots(1)
for i in range(4):
    ax1.scatter(x[y==i,0],x[y==i,1],
               marker = 'o',
               s =5,
                c = colors[i]
               )
plt.show()
```

```python
# 基于上面这个数据我们对KMeans进行分类，首先猜测数据之中有多少簇
from sklearn.cluster import KMeans
n_clusters = 3
cluster = KMeans(n_clusters = n_clusters,random_state =0).fit(x) # 进行训练
y_pred = cluster.fit_predict(x)
y_pred # fit_perdict 显示预测的结果
```

```python
pre = cluster.labels_
pre
```

```python
pre == y_pred # 我们可以看到，cluster.labels_返回的和cluster.fit_predict(x)返回的标签是一样的
```

```python
# 基于轮廓系数来选择n_clusters
# 绘制轮廓系数分布图和聚类后的数据分布图来选择我们最佳的n_clusters
# 知道 每个举出来的类的轮廓系数是多少，还想要一个各个类之间的轮廓系数的对比
# 知道聚类完毕之后图像的分布是什么样的
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples,silhouette_score

import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

for n_clusters in range(2,8):
    n_clusters = n_clusters
    # 创建画布，利用这种方式创建2个子图（1,2）1行2列 
    fig,(ax1,ax2) = plt.subplots(1,2)
    # 设置子图的英尺尺寸
    fig.set_size_inches(18,7)
    # 首先我们来设定横坐标轴
    # 轮廓系数的取值范围【-1，1】 之间，但我们至少是希望轮廓系数要大于0的
    # 太长的横坐标不利于我们的可视化，所以这里设定的x轴取值在[-0.1,1]
    ax1.set_xlim([-0.1,1])
    """
    接下来设定纵坐标：通常来说，纵坐标是从0开始的，最大值渠道X.shape[0]的去痣
    但是我们希望，每个簇能够排在一起，不同的簇之间能够有一定的间隙
    以便于我们看到不同的条形图聚在一起的块，理解他们是哪一个簇的
    因此我们在设定纵坐标的取值范围的时候，在X.shap[0]上加上一个距离(n_clusters+1)*10做间隙
    """
    ax1.set_ylim([0,x.shape[0]+(n_clusters +1)*10])
    # 开始建模，调用聚类好的标签
    clusterer = KMeans(n_clusters = n_clusters,random_state = 10).fit(x)
    # 调用标签
    cluster_labels = clusterer.labels_
    # 调用轮廓系数分数，注意，silhouette_score生成的是所有样本店的轮廓系数的均值
    # 需要2个输入的参数：特征矩阵X和聚类完毕后的标签
    silhouette_avg = silhouette_score(x,cluster_labels)
    #　打印出来结果现在的簇量下，整体轮廓系数究竟是多少
    print("for n_clusters =",n_clusters,
          "The average silhouette_score is:",silhouette_avg)
    # 调用silhouette_samples，返回每一个样本点的轮廓系数，这就是我们的横坐标
    sample_silhouette_values = silhouette_samples(x,cluster_labels)
    
    # 设定y轴的初始距离
    y_lower = 10
    # 接下来对每一个簇进行循环
    for i in range(n_clusters):
        # 从每个样本的轮廓系数结果中抽出第i个簇的轮廓系数，并对它们进行排序
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
        # 注意，sort（）这个命令会直接改掉原数据的循序
        ith_cluster_silhouette_values.sort()
        # 查看这个簇中额度究竟有多少个样本
        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        # 在这个簇y轴上的去痣，应该是由初始值（y_lower）开始，到初始值加上这个簇中的样本数量结束
        y_upper = y_lower+size_cluster_i
        
        """
        colormap的库中，使用小数来调用颜色的函数
        在nipy_spectart([输入任意小数来代表一个颜色])
        在这里我们希望每个簇中的颜色是不同的，我们需要的颜色种类刚好是循环的个数的种类
        在这里，只要保证每次循环生成的小数是不同的，我们可以使用任意方式来获取小数
        在这里我们用的i的浮点数除以n_clusters，在不同的i下，自然生成不同的小数
        以确保所有的簇会有不同的颜色
        """
        color = cm.nipy_spectral(float(i)/n_clusters)
        
        """
        开始填充子图中的内容
        fill_between是让一个范围中的柱状体都统一颜色的函数
        fill_betweenx的范围是在纵坐标轴上
        fill_betweeny的范围是在横坐标轴上
        fill_between下的参数应该输入（纵坐标的下限，纵坐标的上限，x轴上的取值，柱状图的颜色）
        """
        ax1.fill_betweenx(np.arange(y_lower,y_upper),
                         ith_cluster_silhouette_values,
                          facecolor = color,
                          alpha = 0.7
                         )
        # 为每个簇的轮廓系数写上簇的编号，并且让簇的编号显示在坐标轴上的每个条形图的中间位置
        # text参数为（要显示的编号的位置的横坐标，要显示的编号的位置的纵坐标，要显示的编号内容）
        ax1.text(-0.05,y_lower + 0.5*size_cluster_i,str(i))
        # 为下一个簇计算新的y轴上的初始值，是每一次迭代之后，y的上线再加上10
        # 以此来保证，不同的簇的图像之间显示有空隙
        y_lower = y_upper +10
        
    ax1.set_title("This silhouette plot for the various clusters")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("cluster label")
    ax1.axvline(x = silhouette_avg,color = "red",linestyle = '--')
    ax1.set_yticks([])
    ax1.set_xticks([-0.1,0,0.2,0.4,0.6,0.8,1])
    
    colors = cm.nipy_spectral(cluster_labels.astype(float)/n_clusters)
    
    ax2.scatter(x[:,0],x[:,1],
               marker = 'o',
               s = 5,
                c =colors
               )
    centers = clusterer.cluster_centers_
    
    ax2.scatter(centers[:,0],centers[:,1],marker = 'x',c = 'red',alpha =1,s =200)
    
    ax2.set_title("The visualization of the cluster data")
    ax2.set_xlabel("feature sapce for the 1st feature")
    ax2.set_ylabel('feature space for the 2nd feature')
    
    plt.suptitle(("silhouette analysis for Kmeans clustering on sample data \
                  with n_cluster = %d "%n_clusters), fontsize =14,fontweight = 'bold')
    plt.show()
```
