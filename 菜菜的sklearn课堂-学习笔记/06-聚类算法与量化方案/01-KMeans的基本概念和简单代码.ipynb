```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# 创建自己的数据集
x,y = make_blobs(n_samples = 1000,n_features=2,centers =4,random_state =4)

# 创建画布
fig,ax1 = plt.subplots(1)

ax1 = plt.scatter(x[:,0],x[:,1],s = 0.6,marker = 'o')
plt.show()
```

### 如何衡量聚类算法的效果？
> 聚类模型的结果不是某种标签输出，并且聚类的结果是不确定的，其优劣由业务需求或者算法需求来决定，并且没有永远的正确答案

> KMeans的目标是保证“簇内差异小，簇间差异大”，我们就可以通过衡量簇内的差异来衡量聚类的效果

- Inertia是用距离来衡量簇内差异的指标，因此，我们初步使用Inertia来作为聚类的衡量指标，但是这个指标是无界的，**Inertia是越小越好**，但是他会收到特征数目的影响，数据维度很大的时候，Inertia的计算量会陷入维度诅咒之中，计算量会爆炸，不适合一次次的评估模型
- Inertia受超参数的影响，随着K值的增大，Inertia注定会越来越小，但是并不能代表模型效果变好了


### 当真实标签位置的时候：轮廓系数
a：样本同同簇中其他样本的相似度，等于样本与同一簇中所有其他点之间的平均距离
b：样本与其他簇的样本相似度，等于样本与下一个最近的簇中的所有的点之间的平均距离
b > a
单样本的轮廓系数计算为： s = （b-a）/ max(a,b)

所以的轮廓的范围是（-1，1），越接近1的轮廓系数效果越好

- 在sklearn中，我们使用模块metrics中的类silhouette_score来计算轮廓系数：它返回的是一个数据集中，所有样本的轮廓系数的均值
- 在metrics模块中的silhouette_sample,他的参数与轮廓系数一样，但是返回的数据集中每个样本自己的轮廓系数

```python
from sklearn.metrics import silhouette_score
from sklearn.metrics import silhouette_samples

silhouette_score(x,y_pred)
```

```python
silhouette_score(x,cluster_.labels_)
```

```python
silhouette_samples(x,y_pred)
```

### 当真实标签未知的时候：卡林斯基-哈拉巴斯指数
- 哈拉巴斯指数越高越好
- 虽然calinsiki_harabasz_score指数没有边界，但是比起轮廓系数，他有一个巨大的优点，就是计算速度十分的快，可以使用时间戳进行计算运行时间
------------

```python
from sklearn.metrics import calinski_harabasz_score
calinski_harabasz_score(x,y_pred)
```

```python
from time import time
t0 = time()

calinski_harabasz_score(x,y_pred)
print(time()-t0)

t1 = time()
silhouette_score(x,y_pred)
print(time()- t1)
```
